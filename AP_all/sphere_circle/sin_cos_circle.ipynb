```python
# vary the theta from 0 -> 2pi and plot y as cos(theta) and x as sin(theta)
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
step = 0.1
x= []
y= []
```

```python
for i in np.arange(0,round(2*np.pi,5)+step,step):
    x.append(np.sin(i))
    y.append(np.cos(i))
    # np.sin(rads)
    
```

```python
plt.plot(x,y,marker='o')
plt.title("circle using sin cos")
plt.xlabel("sin(i)")
plt.ylabel("cos(i)")
plt.grid(True)
plt.gca().set_aspect(1) # means the scaling of y axis = 1* x scaling of x axis
plt.show()
```
